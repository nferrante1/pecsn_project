```python
# Funzioni per prelevare Utente, ID attività da file di testo CSV
def get_user_activity_from_filename(activity_file_name):
    if not (activity_file_name[0:6]=='HuGaDB'): 
        return 0 #non é un dataset
    if (activity_file_name[-9]=='0'):
        user_id=activity_file_name[-8]
    else:
        user_id=activity_file_name[-9:-7]
    #Se l'utente inizia con zero rimuovere lo zero
    activity=activity_file_name[10:-10]
    return [user_id,activity]
```

```python
def to_intStandardUnits(dataset):
    #Funzione per portare il dataset HuGaDB in misure standard internazionali
    for (columnName, columnData) in dataset.iteritems():
        if 'acc' in columnName:
            #Converto accelerazione in m/s^2
            dataset[columnName]=dataset[columnName]*(9.80665/32768)
        if 'gyro' in columnName:
            #Converto giroscopio in rad/sec
            dataset[columnName]=dataset[columnName]*(2000/32768)
        if 'EMG' in columnName:
            #Converto EMG in Volts
            dataset[columnName]=dataset[columnName]*(0.001/255)
```

```python
# Imposto le directory che mi serviranno per il nuovo dataset

directory="./HuGaDB_Data/"
dir_intStandardData="./data/labelled_data/"
dir_features="./data/features/"
dir_extractedFeatures="./data/features/extracted_features/"
```

```python
# Creo la cartella per il dataset in misure standard internazionali
if not(os.path.exists('./data')):
    os.mkdir('data')
if not(os.path.exists(dir_intStandardData)):
    os.mkdir(dir_intStandardData)
if not(os.path.exists(dir_features)):
    os.mkdir(dir_features)
if not(os.path.exists(dir_extractedFeatures)):
    os.mkdir(dir_extractedFeatures)

# Ottengo tutti gli id degli utenti
user_id_set=set()
# Iterando i file e analizzando il filename grazie alla struttura dello stesso
for filename in os.listdir(directory):    
    print (filename)
    if get_user_activity_from_filename(filename)==0:
        continue
    [us_id,activity]=get_user_activity_from_filename(filename)
    if activity in activities_final and filename.endswith(".txt"):
        user_id_set.add(us_id)
    
# Creo le cartelle per ogni utente
for user_id in user_id_set:
    if not(os.path.exists(dir_intStandardData+"User"+user_id)):
        os.mkdir(dir_intStandardData+"User"+user_id)

print(user_id_set)

```

```python
for filename in os.listdir(directory):
    
    if get_user_activity_from_filename(filename)==0:
        continue
    
    [user_id,activity]=get_user_activity_from_filename(filename)
    
    if activity in activities_final and filename.endswith(".txt"):
        
        
        #Per ogni file carica il dataset
        dataset=pd.read_csv("./HuGaDB_Data/"+filename,skiprows=3,delimiter='\t')
        
#         #Setto in nome degli indici in sample
#         dataset.index.name='Sample'
        
        #Converto in unita' standard
        to_intStandardUnits(dataset)
        
        #Setto in nuovo filename con estensione csv
        filename_csv=filename.replace(".txt",".csv")
        print (filename_csv)
        dataset.to_csv(dir_intStandardData+"User"+user_id+"/"+filename_csv)
        
        
        
        
```
